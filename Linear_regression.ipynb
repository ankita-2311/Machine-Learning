```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("Jupyter_Data/homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='*')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='*')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
plt.show()
```

```python
d = pd.read_csv("Jupyter_Data/areas.csv")
d.head(3)
```

```python
df = pd.read_csv("Jupyter_Data/canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income(US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='*')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income(US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='*')
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
plt.show()
```
